# Pan f-statistics

Convert to `plink` format:
```
plink2 --vcf --max-alleles 2 --snps-only --make-bed --maf 0.00 --out pan_chr19
cp ../share/new_pan_chr19.fam pan_chr19.fam
```

In `R`, extract f2-stats, load the data and calculate f3 for the chimp subspecies compared to bonobo:

```
extract_f2(pref="pan_chr19",
    outdir="genos4",blgsize=500000,
    overwrite=T,maxmem=500)
    
f2_blocks = f2_from_precomp("genos4")

f3_tableP<-f3(f2_blocks,pop1="PPA",
          pop2=c("PTT","PTV","PTE","PTS"),
          pop3="test")

barplot(f3_tableP$est,names.arg=f3_tableP$pop2)
```
